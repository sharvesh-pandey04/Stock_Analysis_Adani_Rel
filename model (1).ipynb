```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
reliance = pd.read_csv("reliance_stock.csv")
```

```python
adani = pd.read_csv("adani_power_stock.csv")
```

```python
reliance.head()
```

```python
adani.head()
```

```python
reliance.reset_index(inplace=True)
```

```python
adani.reset_index(inplace=True)
```

```python
reliance['Date'] = pd.to_datetime(reliance['Date'])
```

```python
adani['Date'] = pd.to_datetime(adani['Date'])
```

```python
reliance = reliance.sort_values(by='Date')
```

```python
adani = adani.sort_values(by='Date')
```

```python
df = pd.merge(reliance, adani, on="Date")
```

```python
corr = df['Close_x'].corr(df['Close_y'])
```

```python
print("Relation (Correlation):", corr)
```

```python
plt.plot(df['Date'], df['Close_x'])
```

```python
plt.plot(df['Date'], df['Close_y'])
```

```python
plt.title("Reliance vs Adani")
```

```python
plt.show()
```

```python
print("Reliance Mean:", reliance['Close'].mean())
```

```python
print("Reliance Median:", reliance['Close'].median())
```

```python
print("Reliance Std Dev:", reliance['Close'].std())
```

```python
print("Adani Mean:", adani['Close'].mean())
```

```python
print("Adani Median:", adani['Close'].median())
```

```python
print("Adani Std Dev:", adani['Close'].std())
```

```python
rel = np.corrcoef(reliance['Close'], adani['Close'])
```

```python
print("Relation Matrix:\n", rel)
```
